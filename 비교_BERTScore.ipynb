```python
import pandas as pd
import glob
import os

def load_and_merge_csv_no_header(folder_path):
    csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
    df_list = [pd.read_csv(file, header=None) for file in csv_files]
    merged_df = pd.concat(df_list, ignore_index=True)
    return merged_df

# 폴더별 통합
df_deepchat = load_and_merge_csv_no_header("result_deepchat")
df_deepchat_no_rag = load_and_merge_csv_no_header("result_deepchat_no_rag")

print(df_deepchat.shape)
print(df_deepchat_no_rag.shape)
```

#### RAG를 했을 때 BERTscore값

```python
# 컬럼 "4" 전체를 숫자로 변환 후 평균
mean_column4 = pd.to_numeric(df_deepchat[4], errors='coerce').mean()
print("컬럼 4 전체 평균:", mean_column4)
```

#### RAG를 하지 않았을 때 BERTscore 값

```python
# 컬럼 "4" 전체를 숫자로 변환 후 평균
mean_column4 = pd.to_numeric(df_deepchat_no_rag[4], errors='coerce').mean()
print("컬럼 4 전체 평균:", mean_column4)
```
